```python
import re
import pandas as pd
pd.options.display.max_seq_items = 4000
pd.options.display.max_rows = 4000
pd.options.display.max_columns = 4000
pd.set_option('display.max_rows', 1000)
```

```python
filename = 'fake_job_postings.csv'
job_df = pd.read_csv(filename)
```

```python
job_df.groupby(job_df.fraudulent).count()
```

```python
job_df.department.fillna('unspecified', inplace = True)
job_df.required_education.fillna('Unspecified', inplace = True)
job_df.required_experience.fillna('Unspecified', inplace = True)
job_df.industry.fillna('Unspecified', inplace = True)
job_df.company_profile.fillna('Unspecified', inplace = True)
```

```python
job_df['has_salary_range'] = job_df.salary_range.notnull().astype('int') 
```

```python
job_df.drop('salary_range', axis = 1, inplace = True)
```

```python
pd.DataFrame(job_df.isnull().sum()).T
```

```python
df = job_df.dropna(axis = 0)
df.shape
```

```python
df.groupby(df.fraudulent).count()
```

```python
job_df.industry.value_counts()
```

```python
df.title.value_counts()
```

```python
df.location = df.location.replace(to_replace='[A-Za-z]+,|,|^[\s]', value='', regex=True)
```

```python
df.location = df.location.replace(to_replace='\d+', value='', regex=True)
```
