```python
import re
import pandas as pd
pd.options.display.max_seq_items = 4000
pd.options.display.max_rows = 4000
pd.options.display.max_columns = 4000
pd.set_option('display.max_rows', 1000)
```

```python
filename = 'fake_job_postings.csv'
job_df = pd.read_csv(filename)
```

```python
major_cities = ['Amsterdam', 'Athens', 'Atlanta', 'Auckland', 'Austin', 'Bakersfield', 'Berlin', 'Birmingham', 'Boston',
'Brooklyn', 'Brussels', 'Cambridge', 'Chennai', 'Chicago', 'Cincinnati', 'Dallas', 'Dublin', 'Hampton', 
'Hong Kong', 'Houston', 'Irvine', 'Las Vegas', 'London', 'Los Angeles', 'Melbourne', 'New York', 
'New York City', 'Paris', 'Philadelphia', 'Phoenix', 'Portland', 'Redlands', 'San Diego', 'San Francisco', 
'San Jose', 'San Mateo', 'Santa Monica', 'Seattle', 'St. Louis', 'Stockholm', 'Sydney','Thessaloniki', 
'Toronto', 'Vancouver', 'Virginia Beach','Washington']
```

```python
job_df.location = job_df.location.replace(to_replace='[A-Za-z]+,|,|^[\s]', value='', regex=True)
job_df.location = job_df.location.replace(to_replace='\d+|(\s\s)', value='', regex=True)
```

```python
job_df.groupby(job_df.fraudulent).count()
```

```python
job_df.department.fillna('unspecified', inplace = True)
job_df.required_education.fillna('Unspecified', inplace = True)
job_df.required_experience.fillna('Unspecified', inplace = True)
job_df.industry.fillna('Unspecified', inplace = True)
job_df.company_profile.fillna('Unspecified', inplace = True)
```

```python
job_df['has_salary_range'] = job_df.salary_range.notnull().astype('int') 
```

```python
job_df.drop('salary_range', axis = 1, inplace = True)
```

```python
pd.DataFrame(job_df.isnull().sum()).T
```

```python
df = job_df.dropna(axis = 0)
df.shape
```

```python
df.groupby(df.fraudulent).count()
```

```python
df.location = df.location.replace(to_replace='\d+|(\s\s)', value='', regex=True)
```

```python
vc = df.location.value_counts()
vc = pd.DataFrame(vc)
vc.sort_values("location", 
               ascending = False)
```

```python
mj = vc[vc.location >20]
major_cities = mj.index
major_cities = major_cities.to_list()
```

```python
major_cities = major_cities[2:]
major_cities
```
